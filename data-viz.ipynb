```python
covid = pd.read_csv("data-sets/Latest Covid-19 India Status.csv")
covid.head()
```

```python
fig = plt.figure(dpi=100)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.stem(covid['Active'].values,use_line_collection = True,linefmt="--")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
markerline,stemlines,baselines = axes[0].stem(covid["Active"],covid['Total Cases'],linefmt="#ffac80",markerfmt="o",bottom=1.1,use_line_collection=True)
axes[0].set_yscale("log")
markerline.set_markerfacecolor("none")
axes[0].set_title("active and total cases")

pcm = axes[1].pcolormesh(covid.drop("State/UTs",axis=1),cmap="rainbow")
fig.colorbar(pcm)
axes[1].set_title("all data")
```

```python
fig,ax = plt.subplots(1,3,sharex=True,sharey=True,dpi=100)
min_death = covid["Deaths"].min()
ax[0].plot(covid["Active"],covid['Deaths'],"o")
ax[0].set_title("active and deaths")
ax[1].fill_between(covid['Deaths'],covid["Active"],alpha=0.5,facecolor="#00ffbf")
ax[1].set_yscale("linear")
ax[1].set_title("death and active")

ax[2].hist(covid["Active"],color="red")
ax[2].set_title("active")

ax[0].set_yscale("log")
ax[0].set_xscale("log")
```
