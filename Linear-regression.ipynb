```python
data = pd.read_csv("/content/sample_data/EPL_20_21.csv")
data.head()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
ax[0].bar(data['Matches'],data['Age'])
ax[0].set_xlabel("Matches")
ax[0].set_ylabel("Age")
ax[0].set_xlim(0,38)
ax[0].set_ylim(16,38)

ax[1].bar(data['Penalty_Attempted'],data['Age'])
ax[1].set_xlabel("No.of penalty")
ax[1].set_ylabel("Age")
ax[1].set_xlim(0,10)
ax[1].set_ylim(16,38)
```

```python
data_pivot = pd.pivot_table(data = data , values = ['Age','Matches','Goals','Penalty_Attempted'],index = 'Name')
sns.pairplot(data_pivot)
```
